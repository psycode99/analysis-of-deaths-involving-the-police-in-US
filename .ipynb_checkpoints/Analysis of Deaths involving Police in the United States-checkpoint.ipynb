<h4> Adding a Year column </h4>

```python
sep_vals = df.date.str.split('-', expand=True)
year = pd.to_numeric(sep_vals[0])
df['year'] = year
```

# Number of Deaths Per Year

```python
no_deaths_per_year = df.groupby('year').agg({'name': pd.Series.count})
no_deaths_per_year = no_deaths_per_year.rename(columns={'name':'deaths'})
no_deaths_per_year
```

```python
plt.figure(figsize=(16,8), dpi=200)
plt.title('Number of Deaths per Year', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(2015, 2023, step=1), 
           fontsize=14, 
           rotation=45)
plt.scatter(x=no_deaths_per_year.index, y=no_deaths_per_year.deaths, c='dodgerblue', alpha=0.4, s=100)
plt.plot(no_deaths_per_year.index, no_deaths_per_year.deaths, color='crimson', linewidth=2 )
plt.show()
```

```python
fig = px.bar(x=no_deaths_per_year.index,
            y=no_deaths_per_year.deaths,
            color=no_deaths_per_year.index,
             color_continuous_scale='Aggrnyl',
            title='Number of Deaths per Year (2015-2022)'
            )
fig.update_layout(coloraxis_showscale=False, yaxis_title='Number of deaths', xaxis_title='Year')
fig.show()
```


# Analyzing Manner of Death

```python
manner_of_death = df.groupby('manner_of_death').agg({'manner_of_death':pd.Series.count})
manner_of_death = manner_of_death.rename(columns={'manner_of_death': 'Number'})
manner_of_death
```

```python
fig = px.pie(labels=manner_of_death.index,
            values=manner_of_death.Number,
            names=manner_of_death.index,
            title='Manner of Death of Police shooting victims')
fig.show()
```

# Accumulated Deaths per City

```python
deaths_per_city = df.groupby(['city', 'year'], as_index=False).count()
deaths_per_city = deaths_per_city.sort_values('year')[['year', 'city', 'name']]
deaths_per_city = deaths_per_city.rename(columns={'name':'number of deaths'})
```

```python
cumulative_deaths = deaths_per_city.groupby(by=['city',
                                              'year']).sum().groupby(level=[0]).cumsum()
cumulative_deaths.reset_index(inplace=True)
```

```python
l_chart = px.line(cumulative_deaths,
                  x='year', 
                  y='number of deaths',
                  color='city',
                  hover_name='city',
                 title='Accumulated Number of Deaths per Year for each City')

l_chart.update_layout(xaxis_title='Year',
                      yaxis_title='Accumulated Space Deaths')

l_chart.show()
```
